# Vector fields in cartesian vs pollar coordinates

The relationships between cartesian and polar coordinates are given by 
$$
x = rcos\theta\\
y = rsin\theta
$$

and the inverse by

$$
r = \sqrt{x^2+y^2}\\
\theta = tan^{-1}\frac{y}{x}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

The unit vector $\hat{r}$ is given by setting $r=$, so we have $\hat{r} = cos\theta\hat{x}+sin\theta\hat{y}$.

The $\hat{\theta}$ unit vector, being orthogonal to $\hat{r}$ is given by $\hat{\theta} = -sin\theta\hat{x}+cos\theta\hat{y}$.

Putting these together, the unit vectors transform as follows:

$$
\begin{bmatrix}
\hat{r}\\
\hat{\theta}
\end{bmatrix} 
= 
\begin{bmatrix}
cos\theta & sin\theta\\
-sin\theta & cos\theta
\end{bmatrix}
\begin{bmatrix}
\hat{x}\\
\hat{y}
\end{bmatrix} 
$$

The inverse transform is given by the inverse matrix:

$$
\begin{bmatrix}
\hat{x}\\
\hat{y}
\end{bmatrix}
= 
\begin{bmatrix}
cos\theta & -sin\theta\\
sin\theta & cos\theta
\end{bmatrix}
\begin{bmatrix}
\hat{r}\\
\hat{\theta}
\end{bmatrix} 
$$

```python
theta = np.pi/4


X = [0, np.cos(theta),  np.cos(theta)]
Y = [0, np.sin(theta),  np.sin(theta)]

U = [np.cos(theta), -np.sin(theta), np.cos(theta)]  
V = [np.sin(theta), np.cos(theta), np.sin(theta)]  

circ = plt.Circle((0, 0), radius=1, fill=False)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.add_patch(circ)

plt.quiver(X,Y,U,V, units='xy', angles='xy', scale=1, color=['b', 'r', 'r'])

plt.title(r'$\hat{r}, \hat{\theta}$ unit vectors (red)')

plt.xlim(-2, 2)
plt.ylim(-2, 2)


plt.gca().set_aspect('equal', adjustable='box')

# Show plot with gird
plt.grid()

```

Let us examine how some simple vector fields transform. The first one is

$$
F(r, \theta) = r\hat{r}
$$

By plugging in the above equation for $\hat{r}$, we get

$$
F(r, \theta) = r cos\theta\hat{x} + r sin\theta\hat{y}
$$

Finally, to convert the filed in cartesian coordinates, we use the transformation 
$r = \sqrt{x^2+y^2}$, and we noticed that since $tan\theta = y/x$, 
then $sin\theta = \frac{y}{\sqrt{x^2+y^2}}$ and $cos\theta = \frac{x}{\sqrt{x^2+y^2}}$


$$
F(x, y) = x\hat{x} + y\hat{y}
$$

as expected

```python
# Meshgrid
x, y = np.meshgrid(np.linspace(-5, 5, 10), 
                   np.linspace(-5, 5, 10))
  
# Directional vectors
u = x
v = y
  
# Plotting Vector Field with QUIVER
plt.quiver(x, y, u, v, color='g')
plt.title('Vector Field 1')
  
# Setting x, y boundary limits
plt.xlim(-7, 7)
plt.ylim(-7, 7)
  
# Show plot with gird
plt.grid()
plt.show()
```

Now let's take a vector fields that curls

$$
F(x,y) = y\hat{x} - x\hat{y}
$$

To convert to polar coordinates, we transform the unit vectors fist
$$
F(x, y) = y(cos\theta\hat{r} - sin\theta\hat{\theta}) - x(cos\theta\hat{\theta} + sin\theta\hat{r})
$$
and then the coordinates
$$
F(r, \theta) = rsin\theta(cos\theta\hat{r} - sin\theta\hat{\theta}) - rcos\theta(cos\theta\hat{\theta} + sin\theta\hat{r})
$$
after distribution and cancellation this becomes
$$
F(r, \theta) =  -r\hat{\theta}
$$


```python
# Meshgrid
x, y = np.meshgrid(np.linspace(-5, 5, 10), 
                   np.linspace(-5, 5, 10))
  
# Directional vectors
u = y
v = -x
  
# Plotting Vector Field with QUIVER
plt.quiver(x, y, u, v, color='g')
plt.title('Vector Field 2')
  
# Setting x, y boundary limits
plt.xlim(-7, 7)
plt.ylim(-7, 7)
  
# Show plot with gird
plt.grid()
plt.show()
```
