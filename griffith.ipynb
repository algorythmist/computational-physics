```python
import numpy as np
import matplotlib.pyplot as plt
from functools import partial

def gaussian(x, mean=0, sigma=1):
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mean)/sigma)**2)
```

```python
def plot_function(f, min, max):
    x = np.linspace(min, max,200)
    vf = np.vectorize(f,otypes=[float])
    y = vf(x)
    return plt.plot(x,y)
```

Gaussian Density $\frac{1}{\sigma \sqrt(2\pi)}e^{-(\frac{x-m}{\sigma})^2}$

```python
plot_function(gaussian,-5,5)
```

```python
g_0_5 = partial(gaussian,mean=1, sigma=0.5)
plot_function(g_0_5,-5,5)
g_1_5 = partial(gaussian,mean=1, sigma=1.5)
plot_function(g_1_5,-5,5)
g_2_5 = partial(gaussian,mean=1, sigma=2.5)
plot_function(g_2_5,-5,5)
plt.title("Gaussian Densities")
plt.legend(["sigma =0.5", "sigma = 1.5","sigma = 2.5"])
```

```python
def problem1_4(x, a=1,b=4):
    if x<0 or x>b:
        return 0
    if x <= a:
        return (2*x)/(a*b)
    return (2/b)*(b-x)/(b-a)
```

```python
plot_function(problem1_4,-2,8)
plt.title("Problem 1.4 Density")
plt.axvline(x=1,color='red',label='max')
```
