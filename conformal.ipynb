```python
import numpy as np
import matplotlib.pyplot as plt

xmin=-1.
xmax=1.
ymin=-1.
ymax=1.
step=0.1

for y in np.arange(ymin,ymax,step):
    plt.hlines(y, xmin, xmax, color='red')
for x in np.arange(xmin,xmax,step):
    plt.vlines(x, ymin, ymax, color='blue')
```

```python
for y in np.arange(ymin,ymax,step):
    x=np.arange(xmin,xmax,step)
    z = (x+y*1j)**2
    u = z.real
    v = z.imag
    plt.plot(u, v, color='red')
    
for x in np.arange(xmin,xmax,step):
    y=np.arange(ymin,ymax,step)
    z = (x+y*1j)**2
    u = z.real
    v = z.imag
    plt.plot(u, v, color='blue')
```

```python
for y in np.arange(ymin,ymax,step):
    x=np.arange(xmin,xmax,step)
    z = np.sin(x+y*1j)
    u = z.real
    v = z.imag
    plt.plot(u, v, color='red')
    
for x in np.arange(xmin,xmax,step):
    y=np.arange(ymin,ymax,step)
    z = np.sin(x+y*1j)
    u = z.real
    v = z.imag
    plt.plot(u, v, color='blue')
```

```python
def plot_conformal(fn,excludex=None,excludey=None):
    for y in np.arange(ymin,ymax,step):
        x=np.arange(xmin,xmax,step)
        if excludex and (x>=x[0] or x<=x[1]):
            continue
        z = fn(x+y*1j)
        u = z.real
        v = z.imag
        plt.plot(u, v, color='red')
    
    for x in np.arange(xmin,xmax,step):
        y=np.arange(ymin,ymax,step)
        z = fn(x+y*1j)
        u = z.real
        v = z.imag
        plt.plot(u, v, color='blue')
```

```python
plot_conformal(np.sinh)
```
